```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
import numpy as np
```

# Data import

```python
arima_results = pd.read_csv('arima_results.csv', index_col = 0)
```

```python
arima_results
```

```python
test_data = pd.read_csv('test_data.csv', index_col = 0)
```

```python
test_data
```

# Creating dataframe for results of MAE metric

```python
products = test_data.columns[0:1262]
```

```python
products
```

```python
results = pd.DataFrame(index = products)
```

```python
results["AutoARIMA"] = 0
results["N-BEATS"] = 0
results["Prophet"] = 0
```

```python
results
```

# Processing ARIMA results

```python
type(arima_results.iloc[0,0])
```

```python
arima_results = arima_results.applymap(lambda x: round(x))
```

```python
arima_results
```

```python
mae = []
for product in products:
    y_pred = np.array(arima_results[product])
    y_true = np.array(test_data[product])
    metric = mean_absolute_error(y_true, y_pred)
    mae.append(metric)

results['AutoARIMA'] = mae
results.to_csv('mae_results.csv')
```

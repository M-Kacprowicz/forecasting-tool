```python
import pandas as pd
```

```python
df = pd.read_csv('data_cleaned.csv', index_col = 0)
```

```python
df
```

```python
from datetime import datetime
```

```python
df['Data'] = df.apply(lambda row: datetime.strptime(f"{int(row['Rok'])}-{int(row['Tydzień Roku'])}-1", "%Y-%W-%w"), axis=1)
```

```python
df
```

```python
df.insert(0, 'Data', df.pop('Data'))
```

```python
df
```

```python
df.drop(['Rok', 'Tydzień Roku'], axis = 'columns', inplace = True)
```

```python
df
```

```python
df.to_csv('data_cleaned.csv')
```

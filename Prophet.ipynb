```python
import pandas as pd
from darts import TimeSeries
from darts.models import Prophet
#from darts.utils.holidays import PolishHolidays
```

```python
train_data = pd.read_csv('train_data.csv', index_col = 0)
```

```python
train_data
```

```python
test_data = pd.read_csv('test_data.csv', index_col = 0)
test_data
```

```python
prophet_results = pd.read_csv('prophet_results.csv', index_col = 0)
prophet_results
```

```python
products_train = train_data.columns[0:1262]
```

```python
train_data.index = pd.to_datetime(train_data.index)
test_data.index = pd.to_datetime(test_data.index)
prophet_results.index = pd.to_datetime(prophet_results.index)
```

```python
#PÄtla wykonujÄca prognozowanie dla kaÅ¼dej zmiennej. Trwa ok. 12 minut.
for product in products_train:
    #Konwertowanie DF na TimeSeries z darts
    series = TimeSeries.from_dataframe(train_data, value_cols = product, freq = 'W-MON', fillna_value = 0)
    #Utworzenie modelu
    model = Prophet(country_holidays = 'PL')
    #Dopasowanie modelu do danych
    model.fit(series)
    #Prognoza
    forecast = model.predict(n = len(test_data))
    #Przypisanie wynikÃ³w prognozy
    for x in range(0, len(test_data)):
        prophet_results[product][x] = forecast.pd_dataframe().iloc[x,:][0]

    prophet_results.to_csv('prophet_results.csv')
    
    print(f'Wykonano prognozÄ dla zmiennej {product}.', end = ' ')
```

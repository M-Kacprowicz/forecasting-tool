```python
import pandas as pd
from darts import TimeSeries
from darts.models import NBEATSModel
```

```python
train_data = pd.read_csv('train_data.csv', index_col = 0)
```

```python
train_data
```

```python
test_data = pd.read_csv('test_data.csv', index_col = 0)
```

```python
test_data
```

```python
nbeats_results = pd.read_csv('nbeats_results.csv', index_col = 0)
```

```python
nbeats_results
```

```python
products_train = train_data.columns[0:1262]
```

```python
products_train
```

```python
train_data.index = pd.to_datetime(train_data.index)
test_data.index = pd.to_datetime(test_data.index)
nbeats_results.index = pd.to_datetime(nbeats_results.index)
```

```python
#Pętla wykonująca prognozowanie. Trwa około 11.5 godziny.
for product in products_train:
    #Konwertowanie DF na TimeSeries z darts
    series = TimeSeries.from_dataframe(train_data, 
                                       value_cols = [product, 
                                                     'Cena rudy żelaza', 
                                                     'Stan epidemii w Polsce', 
                                                     'Inflacja'], 
                                       freq = 'W-MON', 
                                       fillna_value = 0)
    #Utworzenie modelu
    model = NBEATSModel(input_chunk_length = len(train_data) - len(test_data), 
                    output_chunk_length = len(test_data), 
                    generic_architecture = True
                   )

    #Dopasowanie modelu do danych
    model.fit(series)
    #Prognoza
    forecast = model.predict(n = len(test_data))
    #Przypisanie wyników prognozy
    for x in range(0, len(test_data)):
        nbeats_results[product][x] = forecast.pd_dataframe().iloc[x,:][0]
    
    print(f'Wykonano prognozę dla zmiennej {product}.')
```

```python
nbeats_results
```

```python
nbeats_results.to_csv('nbeats_results.csv')
```
